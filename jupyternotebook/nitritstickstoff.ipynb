```python
nitrate = pd.read_csv("test.csv")
```

```python
nitrate['datum_pn'] = pd.to_datetime(nitrate['datum_pn'], format = '%Y-%m-%d')
```

```python
nitrate.dtypes
```

```python
import geopandas as gpd
```

```python
nitrate_shp = gpd.GeoDataFrame(nitrate, crs = 'EPSG:25832', geometry = gpd.points_from_xy(nitrate.e32, nitrate.n32))
```

```python
nitrate_shp
```

```python
nitrate_shp = nitrate_shp.drop("geom", axis=1)
```

```python
nitrate_shp
```

```python
data_in_dir = r"../data/shapefiles/"

file_fname = r"nitritstickstoff.gpkg"

file_pfname = data_in_dir + file_fname
```

```python
nitrate_shp.to_file(file_pfname, layer='nitritstickstoff', driver="GPKG")
```
